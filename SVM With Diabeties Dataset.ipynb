```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
df=pd.DataFrame(data)
```

```python
print(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=pd.DataFrame(data['Insulin'])
y=pd.DataFrame(data['Glucose'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#creation of model
from sklearn import linear_model
reg=linear_model.LinearRegression()
```

```python
#train the model using the training sets and also sit the values
reg.fit(x_train,y_train)
```

```python
#make predictions using the testing set
y_pred=reg.predict(x_test)
print(y_pred)
```

```python
#finding values of coefficient 
print('cofficient : \n',reg.coef_)
print('intercept : \n',reg.intercept_)
```

```python
#to find errors
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
print('Mean Squared error : %.2f'%mean_squared_error(y_test,y_pred))
```

```python
print('Mean Absolute error : %.2f'%mean_absolute_error(y_test,y_pred))
```

```python
from sklearn import metrics
print('Root Mean Squared error : %.2f'%np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
#plotting

plt.scatter(x_test,y_test , color='red')
plt.plot(x_test,y_pred, color='blue',linewidth=3)
plt.show()
```
